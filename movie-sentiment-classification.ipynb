```python
import nltk
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
file_path = '/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv'
#movies=pd.read_csv("/kaggle/input/imdb-dataset-of-50k-movie-reviews")
df = pd.read_csv(file_path)
print(df.head())
df.info()
```

```python
# Pobierz stopwords dla języka angielskiego
nltk.download('stopwords')
nltk.download('punkt')
# stop_words = set(stopwords.words('english'))
```

```python
vectorizer=TfidfVectorizer(max_features=500,stop_words="english")
```

```python
X=vectorizer.fit_transform(df.review)
```

```python
X = vectorizer.fit_transform(df['review'])
y = df['sentiment']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
np.unique(y)
```

```python
classifier=SVC(kernel='linear')
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
accuracy=accuracy_score(y_pred,y_test)
classify=classification_report(y_pred,y_test)
confuse=confusion_matrix(y_pred,y_test)
```

```python
print(accuracy)
print(classify)
print(confuse)
```
